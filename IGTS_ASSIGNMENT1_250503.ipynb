```python
import re
def is_palindrome_simplified(s):
    # Use regex to remove all non-alphanumeric characters and convert to lowercase
    cleaned_s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    # Check if the cleaned string is equal to its reverse
    return cleaned_s == cleaned_s[::-1]
```

```python
numbers_str = input("Enter a list of integers separated by commas: ")
numbers = [int(num.strip()) for num in number_str.split(',') if num.strip()]
x = int(input("Enter a number x: "))
```

```python
found_pair = False
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if abs(numbers[i] - numbers[j]) == x:
            print(f"Found pair: ({numbers[i]}, {numbers[j]}) with absolute difference {x}")
            found_pair = True
            break
    if found_pair:
        break
```

```python
import numpy as np
scores = np.array([45, 82, 76, 55, 90, 38, 67, 71])
mean_score = np.mean(scores)
median_score = np.median(scores)
std_dev_score = np.std(scores)

print(f"Mean score: {mean_score:.2f}")
print(f"Median score: {median_score:.2f}")
print(f"Standard deviation of scores: {std_dev_score:.2f}")
min_score = np.min(scores)
max_score = np.max(scores)

normalized_scores = (scores - min_score) / (max_score - min_score)
print("Normalized scores (0 to 1):")
print(normalized_scores)
mean_score_for_indices = np.mean(scores)
indices_above_mean = np.where(scores > mean_score_for_indices)

print(f"Scores above the mean ({mean_score_for_indices:.2f}):")
print(scores[indices_above_mean])
print("Indices of students who scored above the mean:")
print(indices_above_mean[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)

mean_return = 0.0008
std_dev_return = 0.015
daily_returns = np.random.normal(mean_return, std_dev_return, 252)

initial_price = 100
price_series = initial_price * (1 + daily_returns).cumprod()

price_series = np.insert(price_series, 0, initial_price)

print("First 5 daily returns:", daily_returns[:5])
print("First 5 prices in the series:", price_series[:5])
plt.figure(figsize=(12, 6))
plt.plot(price_series)
plt.title('Simulated Price Series from Daily Returns')
plt.xlabel('Days')
plt.ylabel('Price')
plt.grid(True)
plt.show()

final_price = price_series[-1]

peak_price = np.max(price_series)

running_max = np.maximum.accumulate(price_series)
drawdown = (running_max - price_series) / running_max
max_drawdown = np.max(drawdown)

print(f"Final Price: {final_price:.2f}")
print(f"Peak Price: {peak_price:.2f}")
print(f"Maximum Drawdown: {max_drawdown:.2%}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan','Feb','Mar','Apr','May','Jun',
'Jul','Aug','Sep','Oct','Nov','Dec']
year1 = [42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80]
year2 = [52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95]
difference = np.array(year2) - np.array(year1)
colors = ['green' if diff > 0 else 'red' for diff in difference]

fig, axes = plt.subplots(2, 1, figsize=(12, 10))

axes[0].plot(months, year1, label='Y1')
axes[0].plot(months, year2, label='Y2')
axes[0].set_title('Monthly Revenue Comparison')
axes[0].set_xlabel('Months')
axes[0].set_ylabel('Revenue')
axes[0].legend()

axes[1].bar(months, difference, color=colors)
axes[1].set_title('Month-by-Month Revenue Difference (Year 2 - Year 1)')
axes[1].set_xlabel('Months')
axes[1].set_ylabel('Revenue Difference')

plt.tight_layout()
plt.show()
```

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
returns = np.random.multivariate_normal(
mean=[0.0005, 0.0008, 0.0003],
cov=[[0.0004, 0.0001, 0.00005],
[0.0001, 0.0006, 0.00008],
[0.00005, 0.00008, 0.0003]],
size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)

df_prices = pd.DataFrame(prices, columns=['Asset_A', 'Asset_B', 'Asset_C'])

print("Price DataFrame head:")
display(df_prices.head())
plt.figure(figsize=(12, 7))
plt.plot(df_prices['Asset_A'], label='Asset_A')
plt.plot(df_prices['Asset_B'], label='Asset_B')
plt.plot(df_prices['Asset_C'], label='Asset_C')

plt.title('Simulated Asset Price Series')
plt.xlabel('Time (Days)')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

daily_returns_df = df_prices.pct_change().dropna()

correlation_matrix = daily_returns_df.corr()

print("Correlation Matrix of Daily Returns:")
display(correlation_matrix)

portfolio_weights = np.array([1/3, 1/3, 1/3])

portfolio_returns = (daily_returns_df * portfolio_weights).sum(axis=1)

initial_portfolio_value = 100
portfolio_value = initial_portfolio_value * (1 + portfolio_returns).cumprod()

plt.figure(figsize=(12, 7))
plt.plot(df_prices['Asset_A'], label='Asset_A')
plt.plot(df_prices['Asset_B'], label='Asset_B')
plt.plot(df_prices['Asset_C'], label='Asset_C')
plt.plot(portfolio_value, label='Equally-Weighted Portfolio')

plt.title('Simulated Asset Prices and Equally-Weighted Portfolio Value')
plt.xlabel('Time (Days)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
